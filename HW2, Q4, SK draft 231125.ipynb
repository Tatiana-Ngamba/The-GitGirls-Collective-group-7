## How our group managing our code

We share and manage our code via our **GitHub repository** _https://github.com/Tatiana-Ngamba/The-GitGirls-Collective-group-7/tree/main_  
Our thorough practice and learnings from Assignment 4 helped us set a structured but adaptable Repository structure:
* **main** branch remains relatively clean, reserved for final Pull Request(s) with mature files
* communal working branch 1: **Project_Code**
* communal working branch 2: **Project_Documentation**
* smaller branches will be utilised for significant but smaller work stages, like SQL-Database branch and Datasets branch
* **individual members** will create their own named branch from whichever branch holds the files they are working on. Members will create pull requests from their own branch into the 'communal' destination branch when code/file updates are ready for review. This allows each member 'safe spaces' from which to alter and experiment with updating code.

We're not afraid to **adapt** and if, for example, we found too many files in one branch, or realised a subset of files (i.e. raw dataset files) could be better categorised separately, we will create new branches or even use subfolders within branches.  We appreciate the difference between some files being mainly static, and just there for reference (i.e. datasets) whilst other files have dynamic content and will often be changing. 

The group has significantly improved **professionalism and discipline**, keeping control of file-sharing/versioning with GitHub whilst Slack shares links and general communications.
